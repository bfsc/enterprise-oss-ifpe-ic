```python
# Importa as bibliotecas
import pandas as pd
import os
import json
import requests
import whois
```

```python
# Carrega o DataSet
df = pd.read_csv(r"../Tabela_spinelli/enterprise_projects.txt", sep="\t", header=None)
df.columns = ['url', 'project_id', 'sdtc', 'mcpc', 'mcve', 'star_number', 'commit_count', 'files',
                'lines', 'pull_requests', 'most_recent_commit', 'committer_count', 'author_count',
                'dominant_domain', 'dominant_domain_commiter_commits', 'dominant_domain_author_commits',
                'dominant_domain_committers', 'dominant_domain_authors', 'cik', 'fg500', 'sec10k', 'sec20f',
                'project_name', 'owner_login', 'company_name', 'owner_company', 'license']
df
```

```python
company = df["company_name"].dropna()
company.describe()
```

```python
# Coluna dominant_domain
dominio_dominante = df["dominant_domain"].dropna()
dominio_dominante
```

```python
# Faz o request na whoIs
nome_empresas = []
sim = nao = 0
for dominio in dominio_dominante:
    print(f"Analizando o dominio [{dominio}]")
    try:
        w = whois.whois(dominio)
        empresa = w["org"]
    except Exception as e:
        nome_empresas.append("NaN")
        print("-" * 10 + f"Falha na Análise: {e}")
        nao += 1
    else:
        if empresa is None:
            nome_empresas.append("NaN")
            print("-" * 10 + f"Falha na Análise")
            nao += 1
        else:
            nome_empresas.append(empresa)
            print("-" * 10 + "Analizado com Sucesso!")
            sim += 1
        
        
```

```python
print(nome_empresas)
```

```python
# Salva os nomes em um arquivo txt
with open("company_names.txt", "w", encoding="utf-8") as arquivo:
    for nome in nome_empresas:
        arquivo.write(f"{nome}\n")
```

```python
# Adiciona uma nova coluna com os nomes das empresas conseguidas
df2 = df.assign(company_name_whois = nome_empresas)

df2
```

```python
# Gera uma tabela excel com a nova tabela
df2.to_excel("Tabela_Spinellis_whois.xlsx")
```
