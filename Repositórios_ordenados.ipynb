```python
import pandas as pd
import openpyxl
import json
import requests
from time import sleep
from random import randint
```

```python
df.to_excel('planilhas/enterprise_projects.xlsx', 'Sheet1', index=False, header=False)
```

Pega o owner_login e o project_name e geram uma lista de URL's para os respectivos arquivos Json

```python
urls = []
```

```python
book = openpyxl.load_workbook('planilhas/enterprise_projects_column_names.xlsx')
first_page = book['Sheet1']
for rows in first_page.iter_rows(min_row=2):
    urls.append(f'https://api.github.com/repos/{rows[23].value}/{rows[22].value}')
print(urls)
```

Verifica os tamanhos dos repositrios no analisados 

```python
tamanhos_repositorios = {}
file = open('urls_e_tamanhos_conseguidos.txt', 'r')
read_file = file.read()
c = 0
for url in urls:
    if c % 100 == 0:
        print(c)
    if url not in read_file:
        try:
            sleep(randint(0, 10))
            texto = requests.get(url=url)
            data = json.loads(texto.text)
            if 'size' in data:
                tamanhos_repositorios[url] = data['size']
        except Exception as e:
            print(e)
        c += 1
file.close()
print(c)
```

```python
print(tamanhos_repositorios)
```

Verifica a quantidade de itens dentro do dicionrio e ordena os tamanhos em ordem crescente

```python
cont = 0
for v in tamanhos_repositorios:
    cont += 1
print(cont)
```
